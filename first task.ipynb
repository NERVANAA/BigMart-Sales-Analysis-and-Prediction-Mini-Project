#Data Cleaning



```python
print(df.dtypes)
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
df = df.drop(['id', 'Unnamed: 0.1','first_investment_at','last_investment_at', 'short_description','description'], axis=1)
df
```

```python
categorical = ['name' ,'normalized_name',"permalink" , 'homepage_url', 'logo_url','overview','tag_list','domain']
for cat in categorical:
  df[cat].fillna('unnkown',inplace=True)
```

```python
date = ['founded_at', 'first_funding_at', 'last_funding_at', 'first_milestone_at', 'last_milestone_at']

df[date] = df[date].fillna(method='ffill')
```

```python
num_columns1 = ['logo_width','logo_height','funding_rounds','funding_total_usd','milestones','relationships','ROI']

for col6 in num_columns1:
    mean_val = df[col6].mean()
    df[col6] = df[col6].fillna(mean_val)
```

```python

cols_to_fill = ['entity_type', 'category_code', 'status', 'founded_at',
                'country_code', 'state_code', 'city', 'created_by']
df[cols_to_fill] = df[cols_to_fill].fillna(df.mode().iloc[0])
```

```python
df = df.drop(['entity_id', 'parent_id','investment_rounds','invested_companies'], axis=1)
df
```

```python
df = df.drop(['closed_at', 'twitter_username'], axis=1)
df
```

```python

num_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(num_cols)
```

```python
df['lng'] = df['lng'].interpolate(method='linear', limit_direction='both', axis=0)
df['lat'] = df['lat'].interpolate(method='linear', limit_direction='both', axis=0)
```

```python
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
for col in df.columns:
    if df[col].dtype == 'datetime64[ns]':
        df[[f'{col}_Day', f'{col}_Month', f'{col}_Year']] = df[col].apply(lambda x: pd.Series([x.day, x.month, x.year]))
```

```python

numeric_columns = ['logo_width','logo_height','funding_rounds','funding_total_usd','milestones',
                   'relationships','lat','lng','ROI']

# Selecting only the numeric columns and creating a new dataframe
num_col_df = df.select_dtypes(include=['int', 'float'])

# Describing the numeric columns of the dataframe
num_col_df.describe()
```

```python

# Find columns with outliers using z-score
outlier_cols = []
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        z_scores = np.abs((df[col] - df[col].mean()) / df[col].std())
        if z_scores.max() > 3:
            outlier_cols.append(col)

print(f"Columns with outliers: {outlier_cols}")
```

```python
cols=['logo_width', 'logo_height', 'funding_rounds', 'funding_total_usd', 'milestones', 'relationships', 'lat', 'lng', 'ROI']
for col in cols:
        Q1 = df[col].quantile(.25)
        Q3 = df[col].quantile(.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outli = (df[col] < lower) | (df[col] > upper)
        df.loc[outli,col]=df[col].median()
```

```python
#exploring the distibution of variable through histogram or box plots
df.hist(figsize=(10,10))
plt.show()
df.plot(kind="box",subplots=True, layout=(10,10),figsize=(20,20))
plt.show()
```
